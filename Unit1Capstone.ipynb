```python
df.rename(columns={'Year':'Year', 'Average Life Expectancy (Years)':'Average_Life_Expectancy', 'Age-adjusted Death Rate':'Adjusted_Death_Rate',\
                        'National_Health_Expense':'National_Health_Expense', 'Average_Annual_Change':'Average_Annual_Change'},\
               inplace=True)
```

```python
sns.catplot(data=df, x='Year', y='National_Health_Expense', kind='swarm')
sns.catplot(data=df, x='Year', y='Average_Life_Expectancy', kind='swarm')
```

When comparing the two charts it can be seen that there is a jump in average life expectancy(Average_Life_Expectancy) between the years of 2003-2004, while the national health expense(National_Health_Expense) follows a linear path. The comparison differs as the tail end of average life expectancy does not continue to grow, while the national health expense continues to climb.

```python
sns.lmplot(data=df, x='Average_Life_Expectancy', y='National_Health_Expense', fit_reg=False)
```

Perhaps a certain amount of expenditure for healthcare can increase the average life expectancy. While looking at the 75% range of average life expectancy it shows that it clusters near the 50%% range of cost association. This may show that there are some indications that more expensive healthcare can boost life expectancy. There are many factors associated with life expectancy so this requires a comparison to see if there are any statistical significance between these two variables.

---

Conducting the T-Test between the two variables (average life expectancy and national health expense)

```python
sns.heatmap(df.corr(), linewidths=0.25, square=True, cmap = "BrBG", linecolor='black', annot=True)
```

```python
ttest_ind(df['Average_Life_Expectancy'], df['National_Health_Expense'])
```

Looking at the above data it shows that using the pvalue is 4.7 which does not reject the null hypothesis.  This means that there is no significant differences between life expectancy and national health expense.  This means that even though there might be an inverse relationship as seen in the heat map there is not enough to statistical significance occuring between the two data sets to differentiate the populations associated with each data set.
