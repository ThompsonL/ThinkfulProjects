```python
x_ar = [15]
for i in range(500):
    y = 5 + .7*x_ar[i-1] 
    x_ar.append(float(y))
```

```python
x_ar = [15]
for i in range (500):
    y = 10 + .25*x_ar[i-1]
    x_ar.append(float(y))
```

```python
plt.plot(x_ar)
plt.title('Autoregressive Example')
plt.show()
```

```python
x_ma = [15]
e_t = 0
for i in range(500):
    e = np.random.randn(1)
    y = 1 + .3*e_t + e
    x_ma.append(float(y))
    e_t=e
```

```python
plt.plot(x_ma)
plt.title('Moving Average Example')
plt.show()
```
